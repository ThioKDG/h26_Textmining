#### 3-2. 데이터 프레임의 열(컬럼) 추가

```python
score_dict = {
    '이름' : ['정약용', '이순신',  '이율곡', '홍길동'],
    '국어' : [70, 100, 90, 85],
    '영어' : [80, 80, 95, 100],
    '수학' : [100, 70, 90, 90]
}   

score_df = pd.DataFrame(score_dict)
score_df
```

```python
score_df['사회'] = [85, 88, 92, 96]
score_df
```

```python
score_df.사회 = 100
score_df
```

```python
# 이미 데이터가 있는 필드는 '.컬럼명' 이런 식으로 작성이 가능 다만 신규 작성시는 예외
score_df['총점'] = score_df.국어 + score_df['수학'] + score_df['영어'] + score_df['사회']
score_df['평균'] = score_df.총점 / 4 
score_df
```

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]
df = pd.DataFrame()
df['x'] = x
df['y'] = y
df['x * y'] = df.x * df.y
df
```

#### 3-4. 데이터 프레임의 행/열 삭제, 인덱스 초기화

```python
score_dict = {
    '이름' : ['정약용', '이순신',  '이율곡', '홍길동'],
    '국어' : [70, 100, 90, 85],
    '영어' : [80, 80, 95, 100],
    '수학' : [100, 70, 90, 90]
}   

score_df = pd.DataFrame(score_dict)
score_df
```

```python
score_df.drop(['이름'], axis=1)
```
