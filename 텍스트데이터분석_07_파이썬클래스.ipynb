```python
class Account: # 클래스의 시작은 Pascal스타일로
    # balance = 100
    def __init__(self): # 멤버 변수 balance 초기화 -> 변수 선언
        self.__balance = 0
    def deposit(self, amount):
        self.__balance+= amount
    def withdraw(self, amount): # 실제로는 잔액 부족을 확인해야 되는 코드 필요
        self.__balance -= amount
    def get_balance(self):
        return self.__balance

a1 = Account()
a1.deposit(100)
a1.withdraw(50)
print('a1 : ',a1.get_balance)
```
