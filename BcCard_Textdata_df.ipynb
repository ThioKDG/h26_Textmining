```python
card = './data/bccard_201906.csv'
card_df = pd.read_csv(card)
# 데이터 확인 앞부분
card_df.head()
```

##### 2. 서울시 거주 / 비거주 고객 비교 분석

```python
# 고객 수
custCount = len(card_df[card_df.CSTMR_MEGA_CTY_NM.isin(['서울특별시'])].value_counts())
nonSeoul = len(card_df[~card_df.CSTMR_MEGA_CTY_NM.isin(['서울특별시'])])
print(f'서울특별시 고객 수 : {custCount:,}')
print(f'서울 비거주 고객 수 : {nonSeoul:,}')
```

```python
# 카드 이용건수
cardCount = card_df.loc[card_df['CSTMR_MEGA_CTY_NM'] == '서울특별시', 'CNT'].sum()
print(f'총 서울특별시 고객 카드 이용 건수 : {cardCount}')
```

```python
# 총 소비액
totalAMT = card_df.loc[card_df['CSTMR_MEGA_CTY_NM'] == '서울특별시', 'AMT'].sum()
print(f'총 서울특별시 고객 총 소비액 : {totalAMT:,} 원')
```

### 편의점 소비액, 서울시 내 편의점 소비액

```python
# 서울시 내 편의점 소비액
seoul_conv_df = card_df[(card_df['MEGA_CTY_NM'] == '서울특별시') & (card_df['TP_BUZ_NO'] == 4010)]
seoul_conv_amt = seoul_conv_df['AMT'].sum()

# 강남구 내 편의점 소비액
gangnam_conv_amt = seoul_conv_df.loc[seoul_conv_df['CTY_RGN_NM'] == '강남구', 'AMT'].sum()

print(f"서울시 전체 편의점 소비액: {seoul_conv_amt:,}원")
print(f'강남구 내 편의점 소비액 : {gangnam_conv_amt:,}원')
```

```python
# 서울시 편의점 데이터(seoul_conv_df)를 기준으로 분류
# 서울 거주자
seoul_amt = seoul_conv_df.loc[seoul_conv_df['CSTMR_MEGA_CTY_NM'] == '서울특별시', 'AMT'].sum()

# 비거주자 (서울이 아닌 경우)
non_seoul_amt = seoul_conv_df.loc[seoul_conv_df['CSTMR_MEGA_CTY_NM'] != '서울특별시', 'AMT'].sum()

print(f"서울 편의점 이용 - 서울 거주자 소비액: {seoul_amt:,}원")
print(f"서울 편의점 이용 - 비거주자 소비액: {non_seoul_amt:,}원")
```
